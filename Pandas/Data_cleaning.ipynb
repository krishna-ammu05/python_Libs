```python
import pandas as pd
import numpy as np
```

# 1. Data Cleaning

```python
df=pd.read_csv('Files/data.csv')
```

```python
#Boolean mask of missing values
df.isnull()
```

```python
#Count of missing values per column
df.isnull().sum()
```

```python
#Boolean mask of non-missing values
df.notnull()
```

```python
#Boolean mask of duplicate rows
df.duplicated()
```

```python
#Remove duplicate rows
df.drop_duplicates()
```

```python
#Remove rows with Missing values
df.dropna()
```

# 2. Fill Missing Values

```python
#Fill Missing Values
df.fillna(0)
```
