# 1. Data Type Conversion & Info

```python
#Convert Column type
#fill missing values with 0 and convert to int
df['quantity'] = df['quantity'].fillna(0).astype(int)

```

```python
#Drop rows with missing values in 'quantity'
df=df.dropna(subset=['quantity'])
df['quantity']=df['quantity'].astype(int)
```

```python
#select columms by type
df.select_dtypes(include=['float64','int64'])
```

```python
df.dtypes
```

```python
#Automatically convert columns to best types
df.convert_dtypes()
```
